```python
listNames = []
listPromedio = []
listInfo1 = []
listMiddle1 = []
listInfo2 = []
listMiddle2 = []
def extractInfo(driver,listNames,listPromedio,listInfo1,listMiddle1,listInfo2,listMiddle2):
    page_content = driver.page_source
    soup = BeautifulSoup(page_content, 'html.parser')

    block= soup.find_all("div", class_="bfdHYd")
    for search in block:
        names = search.find("div", class_= "qBF1Pd")
        data0 = search.find("span",  class_="MW4etd")
        dat1 = search.find_all("div", class_="W4Efsd")
        dat2 = search.find_all("div", class_="ah5Ghc")
        if names:
            ns = names.find("span")
            if ns: 
                print("Nombre:",ns.string)
                listNames.append(ns.string)
        if data0:
            print("puntaje:",data0.string)
            listPromedio.append(data0.string)
        elif data0 == None:
            listPromedio.append(data0)

        if dat1:
            span = dat1[1].find_all("span")
            for text in span:
                if(text.string != None and text.string != '·' ):
                    print("info1:",text.string)
                    listMiddle1.append(text.string)
        listInfo1.append(listMiddle1)
        if dat2:
            for text in dat2:
                span = text.find_all("span")
                for text in span:
                    print("info2:",text.string)
                    listMiddle2.append(text.string)
        listInfo2.append(listMiddle2)

        listMiddle1 = []
        listMiddle2 = []
        print('--------------------------------------------------------------------')
    print("end extract")
```

```python
def extractdata(driver,listNames,listAverage,listAbout,listReview,listDir,listEmail,listPhone,info):
    page_content = driver.page_source
    soup = BeautifulSoup(page_content, 'html.parser')
    names = soup.find("h1", class_= "DUwDvf")
    lugares = soup.find_all("div", class_= "Io6YTe")
    typeB = soup.find("button", class_= "DkEaL")
    block = soup.find_all("div", class_= "F7nice")
    count = 0
    name = ''
    star = 0
    data = ''
    value = []
    if(len(names)==5):
        listAbout.append(typeB.string)
        for search in names:
            if(count == 1):
                name = search.string
                listNames.append(name)
            count = count + 1
        count = 0
        for search in block:
            i = search.find_all("span")
            for re in i:
                if(re.string != None):
                    count = count + 1
                    if(count == 1):
                        star = re.string
                        listAverage.append(star)
                    if(count == 2):
                        data = re.string
                        listReview.append(data)
        count = 0
        for search in lugares:
            if(search.string != None):
                value.append(search.string)
                if(count==0):
                    listDir.append(search.string) 
                if(count==4):
                    listEmail.append(search.string) 
                if(count==5):
                    listPhone.append(search.string) 
                count = count + 1
        print("-------------------------------------------------------------------------------------------------------")
        print(name)
        print(star)
        print(typeB.string)
        print(data)
        print(value)
        info.append(value)
        print("-------------------------------------------------------------------------------------------------------")
```
